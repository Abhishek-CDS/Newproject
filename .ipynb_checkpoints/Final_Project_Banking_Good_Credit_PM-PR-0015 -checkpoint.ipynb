### Insights From table_1.info()
- dt_opened,upload_dt,opened_dt,last_paymt_dt,closed_dt,reporting_dt,paymt_str_dt,paymt_end_dt
  abobe all columns are in object Datatype, but those should be in DateTime datatype.

- customer_no,acct_type,owner_indic,high_credit_amt,cur_balance_amt,amt_past_due,creditlimit,
  cashlimit,rateofinterest,paymentfrequency,actualpaymentamount
  above all columns are in object Datatype, but those should be in Int/Float Datatype.

```python
# To show all columns.
pd.set_option('display.max_columns',None)
table_1.describe(include='O')
```

```python
def change_null(df,col):
    df.loc[df[i]=='',i] = np.nan
```

```python
for i in table_1.columns :
    change_null(table_1,i)
```

```python
date_col = ['dt_opened','upload_dt','opened_dt','last_paymt_dt','closed_dt','reporting_dt',
            'paymt_str_dt','paymt_end_dt']
for i in date_col:
    table_1[i]=table_1[i].astype('datetime64')
```

```python
num_col_1 = ['customer_no','acct_type','owner_indic','high_credit_amt',
             'cur_balance_amt','amt_past_due',
             'creditlimit', 'cashlimit','paymentfrequency','actualpaymentamount']
for i in num_col_1:
    table_1[i]=table_1[i].astype('float32')
```

```python
# To change the value of 'rateofinterest' to convert float

table_1.loc[table_1['rateofinterest']=='"9','rateofinterest'] = 9
table_1.loc[table_1['rateofinterest']=='"8','rateofinterest'] = 8
table_1.loc[table_1['rateofinterest']=='"7','rateofinterest'] = 7
```

```python
table_1.rateofinterest = table_1['rateofinterest'].astype('float32')
```

```python
table_1.rateofinterest.dtype
```

```python
# To show the Statistical information.
table_1.describe()
```

```python
# Saperate Numerecal & catagorical Columns :

numerical_columns = table_1.select_dtypes(include=['int64', 'float32']).columns
categorical_columns = table_1.select_dtypes(include=['object']).columns

print("Numerical Columns:")
print(numerical_columns)
print("\nCategorical Columns:")
print(categorical_columns)
```

```python
# EDA for Numerical Data :
plot1=1
for i in numerical_columns :
    plt.subplot(7,3,plot1)
    sns.histplot(x=table_1[i])
    plot1=plot1+1
plt.tight_layout()
```

```python
# EDA for Catagorical Data. 
plot2=1
for i in categorical_columns.columns :
    plt.subplot(7,3,plot2)
    sns.countplot(x=table_1[i])
    plot2=plot2+1
```
