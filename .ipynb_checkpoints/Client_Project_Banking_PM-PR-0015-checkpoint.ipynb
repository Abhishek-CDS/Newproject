```python
columns_with_same_data = []
# Get the column names of the merged DataFrame

column_names = data.columns

# Use nested loops to compare each column with other columns
for i in range(len(column_names)):
    for j in range(i + 1, len(column_names)):
        column1 = column_names[i]
        column2 = column_names[j]
        
        if data[column1].equals(data[column2]):
            columns_with_same_data.append((column1, column2))

# Display the columns with the same data
print("Columns with the same data:")
for col1, col2 in columns_with_same_data:
    print(f"{col1} and {col2} have the same data.")
```
